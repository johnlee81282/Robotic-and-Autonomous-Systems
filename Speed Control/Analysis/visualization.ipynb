```python
import pandas as pd
import matplotlib.pyplot as plt

test = pd.read_csv("official_mount_test/test_25_10_50.csv")
```

```python
time = 0 - ((test.total_time - test.total_time[0]) / 10000)
plt.plot(time, test.speed / 100)
plt.xlabel("Time (sec)")
plt.ylabel("Speed (ft/sec)")
plt.title("Speed vs Time (kp=25, ki=10)")
plt.show()
```

```python
test = []
# test.append(pd.read_csv("official_mount_test/test_0_0_50.csv"))
# test.append(pd.read_csv("official_mount_test/test_0_5_50.csv"))
# test.append(pd.read_csv("official_mount_test/test_0_10_50.csv"))

# test.append(pd.read_csv("official_mount_test/test_25_0_50.csv"))
# test.append(pd.read_csv("official_mount_test/test_25_5_50.csv"))
# test.append(pd.read_csv("official_mount_test/test_25_10_50.csv"))


test.append(pd.read_csv("official_mount_test/test_50_0_50.csv"))


test.append(pd.read_csv("official_mount_test/test_50_5_50.csv"))


test.append(pd.read_csv("official_mount_test/test_50_10_50.csv"))
```

```python
time = {}
speeds = {}

for i, mount_test in enumerate(test):
    time[i] = 0 - ((mount_test.total_time - mount_test.total_time[0]) / 10000)


for i,  mount_test in enumerate(test):
    speeds[i] = mount_test.speed / 100
```

```python
all_times = [time[key] for key in time]
all_speeds = [speeds[key] for key in speeds]
```

```python
# Loop through each dataset
for t, s in zip(all_times, all_speeds):
    plt.plot(t, s)   # plot one dataset

plt.xlabel("Time (sec)")
plt.ylabel("Speed (ft/sec)")
plt.title("Speed vs Time (kp=50, ki=all)")
plt.legend(['ki=0', 'ki=5', 'ki=10'])  # optional legend
plt.show()
```
